# Importing libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
%matplotlib inline

from sklearn import cluster, datasets
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

import random
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from random import randint
from time import sleep
```

# Importing files

```python
songs_df = pd.read_csv('songs_df.csv')
top_hits = pd.read_csv('top_hits.csv')

pd.set_option('display.max_columns', None)
```

```python
songs_df.head()
```

# split df and scale it

```python
X_cat = songs_df[['song_title','artist']]
```

```python
X = songs_df.select_dtypes(np.number)
X.head()
```

```python
# I will check this one but I'll keep the other one after looking it out

X_scaled = StandardScaler().fit_transform(X)
pd.DataFrame(X_scaled,columns=X.columns).head()
```

```python
# I will use this one in the end for my pipeline

scaler = MinMaxScaler()
X_minmax = scaler.fit_transform(X)
pd.DataFrame(X_minmax,columns=X.columns).head()
```

# checking how many clusters I'll use

```python
K = range(2, 20)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)


plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
K = range(2, 20)
inertia = []

for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_minmax)
    inertia.append(kmeans.inertia_)


plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
%%time

K = range(2, 15)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_scaled)
    silhouette.append(silhouette_score(X_scaled, kmeans.predict(X_scaled)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Silhouette score showing the optimal k')
```

## with the MinMAx Scaler

```python
pca = PCA()
pca.fit(X_minmax)
```

```python
pca.explained_variance_ratio_.round(2)
```

```python

exp_var_pca = pca.explained_variance_ratio_

# Cumulative sum of eigenvalues; This will be used to create step plot
# for visualizing the variance explained by each principal component.

cum_sum_eigenvalues = np.cumsum(exp_var_pca)
#cum_sum_eigenvalues

# Create the visualization plot
plt.bar(range(0,len(exp_var_pca)), exp_var_pca, 
        alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(0,len(cum_sum_eigenvalues)), cum_sum_eigenvalues, 
         where='mid',label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

## Defining the pipeline to get info from spotify

1. input a song
2. check if it is inside top_hits
    * if it is -> recommend another top_hits
3. if it is not a top hit -> search in spotify
4. get the title-artist-uri
5. extract audio features
6. classify them into one of the clusters
7. recommend another song from that cluster

```python
def song_recommender(song):
    # check if it is inside top_hits
    if song in top_hits['title'].values:

        # if it is -> recommend another top_hits
        return print('Maybe this song will also like you! :  ' + random.choice(top_hits['title'].values))

    # if it is not a top hit -> search in spotify    
    else:
        secrets_file = open("secrets.txt","r")
        string = secrets_file.read()
        secrets_dict={}                            
        for line in string.split('\n'):
            secrets_dict[line.split(':')[0]]=line.split(':')[1]

        sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=secrets_dict['cid'],client_secret=secrets_dict['csecret']))

    # getting the info of the song from spotify
        results = sp.search(q=song, type='track', limit=1)
        
    # avoiding problems if it's a bad input    
        if results['tracks']['items'] == []:
            return print("Sorry, we can't find your song :( \nTry with a new input!")
        
        else:
        # get the title-artist-uri
            song = [results['tracks']['items'][0]['name']]
            artist = [results['tracks']['items'][0]['artists'][0]['name']]
            uri = results['tracks']['items'][0]['uri']

        # extract audio features
            audio = sp.audio_features(uri)
            audio = pd.DataFrame.from_dict(audio) 

            # creating df with the info of this song
            df1 = pd.DataFrame({'song_title':song,'artist':artist})
            df = pd.concat([df1, audio], axis=1)

            # separate the numericals to scale
            df_num = df.select_dtypes(np.number)

            # scale the numerical values 
            df_minmax = scaler.transform(df_num)


        # classify them into one of the clusters
            df_cluster = kmeans9.predict(df_minmax)

        # recommend another song from that cluster
            # creating a dataframe with only the songs of this cluster:
            songs_cluster = songs_df2[songs_df2['clusters']==df_cluster[0]]
            #choosing one random song from this group:
            return print('Maybe you will like this song too! :  ' + random.choice(songs_cluster['song_title'].values))

```

```python
# input a song
song = input("What's your favourite song?  ")
song_recommender(song)
```
